```python
##Profitability Metric Calculation
#Create First KPI — Gross Margin %

df["Gross Margin %"] = (df["Gross Profit"] / df["Sales"]) * 100

df[["Sales","Gross Profit","Gross Margin %"]].head()
```

```python
#Pareto Analysis
#What percentage of products generate 80% of total profit?

pareto = (
    df.groupby("Product Name")
    ["Gross Profit"]
    .sum()
    .sort_values(
        ascending=False
    )
    .reset_index()
    )
pareto
```

```python
pareto["Cumulative Profit"] = (
    pareto["Gross Profit"].cumsum()
)

pareto["Cumulative %"] = (
    pareto["Cumulative Profit"] /
    pareto["Gross Profit"].sum()
 ) * 100

pareto
```

```python
#Create a Pareto Chart

import matplotlib.pyplot as plt
plt.Figure(figsize=(13,3))

plt.bar(
    pareto["Product Name"],
    pareto["Gross Profit"]
)

plt.plot(
    pareto["Product Name"],
    pareto["Cumulative %"],
    marker="o"     # letter o
)

plt.xticks(rotation =90)

plt.ylabel("Proft / cumulative %")

plt.title("Pareto Analysis Of Product Profitability")

plt.show()
```

```python
import matplotlib.pylab as plt

plt.figure(figsize=(18,8))

plt.bar(
    pareto["Product Name"],
    pareto["Gross Profit"]
)

plt.xticks(rotation=60, ha="right")

plt.title("Pareto Analysis of Product Profitability")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.barh(
    pareto["Product Name"],
    pareto["Gross Profit"]
)

plt.title("Product Profitability")

plt.xlabel("Gross Profit")

plt.show()
```

```python
#High Sales vs Low Margin

product_summary = (
    df.groupby("Product Name")
    .agg({
        "Sales":"sum",
        "Gross Profit":"sum",
        "Gross Margin %":"mean",
        "Units":"sum"
    })
)

product_summary.sort_values(
    by="Sales",
    ascending=False
)
```

```python
product_summary.sort_values(
    by="Sales",
    ascending=False
)
```

```python
#Let's calculate how much each product costs.

product_summary["Cost"] = (
    product_summary["Sales"]
    - product_summary["Gross Profit"]
 )

product_summary.sort_values (
    by = "Cost",
    ascending=False
)
```

```python
#Create Cost vs Sales Scatter Plot

import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

plt.scatter (
    product_summary["Cost"],
    product_summary["Sales"]
)

top5 = product_summary.nlargest(5, "Gross Profit")

for product in top5.index:
    plt.annotate(
        product,
        (
            top5.loc[product, "Cost"],
            top5.loc[product, "Sales"]
        ),
        fontsize=9
    )

plt.xlabel("Cost"),
plt.ylabel("Sales")
plt.title("Cost Vs Sales by Product")

plt.grid(True)

#plt.tight_layout()

plt.show()
```

```python
product_summary.sort_values(
   by="Cost",
   ascending=False 
)
```

```python
#Monthly Profit Trend

df.groupby(
    pd.Grouper(
        key="Order Date",
        freq="M"
    )
)["Gross Profit"].sum()
```

```python
#Monthly Profit Trend Analysis

monthly_profit=(
    df.groupby(
        pd.Grouper(
            key="Order Date",
            freq="M"
        )
    )["Gross Profit"].sum()
)
monthly_profit
```

```python
#Visualize Monthly Profit Trend

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

monthly_profit.plot()

plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Gross Profit")

plt.grid(True)
plt.show()
```

```python
#Monthly Sales Trend

monthly_sales = (
    df.groupby(
        pd.Grouper(
            key="Order Date",
            freq="M"
        )
    )["Sales"].sum()
)

monthly_sales
```

```python
#Monthly Profit Trend Chart

plt.figure(figsize=(10,6))

monthly_sales.plot()

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.show()
```

```python
#Monthly Margin Trend

monthly_margin = (
    df.groupby(
        pd.Grouper(
            key="Order Date",
            freq="M"
        )
    )["Gross Margin %"].mean()
)

monthly_margin
```

```python
#Monthly Margin Trend Chart

plt.figure(figsize=(10,6))

monthly_margin.plot()

plt.title("Monthly Gross Margin Trend")
plt.xlabel("Month")
plt.ylabel("Gross Margin %")

plt.grid(True)
plt.show()
```

```python
#How much a product's margin changes over time.
#Margin Volatility, standard deviation of margin by product

margin_volatility = (
    df.groupby("Product Name")
    ["Gross Margin %"]
    .std()
    .sort_values(
        ascending=False
                )
    )
margin_volatility
```

```python
#Visualize Margin Volatility

plt.figure(figsize=(12,6))

margin_volatility.plot(kind="barh")


plt.title("Margin Volatility By Product")

plt.xlabel("Standard Deviation of Margin")

#plt.grid(True)

plt.tight_layout()
```
